```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('heart.csv') #reading heart dataset using pandas
```

```python
df
```

```python
df.drop(columns=['gender_number','oldpeak']) #dropping the columns
```

```python
#oldpeaks : oldpeak * 100
```

```python
# adding dropdown and slider options
un_type= df['thal'].unique()
type_dropdown=alt.binding_select(options=un_type)
type_select=alt.selection_single(fields=['thal'],bind=type_dropdown,name='select')
selector=alt.selection_single(name='Select', fields=['age'], init={'age':29},bind=alt.binding_range(min=29,max=77,step=1))
brush = alt.selection_interval()

#Drawing parallel coordinates for the above dataset

alt.Chart(df).transform_window(
    index='count()'
).transform_fold(
    ['rest_bp','serum_chol','max_heartrate','oldpeaks']
).mark_line().encode(
    x='key:N',
    y='value:Q',
    detail='index:N',
    shape='fasting_sugar:N',
    opacity=alt.value(0.5),
    tooltip=['rest_bp','serum_chol','max_heartrate','age','fasting_sugar','sex'],
    color='sex:N'
).add_selection(brush).add_selection(type_select).transform_filter(type_select).add_selection(selector).transform_filter(selector).properties(width=500).interactive()
```
